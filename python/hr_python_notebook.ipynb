### 7) Feature Engineering and Segmentation

```python
# Business-driven experience segmentation for workforce analysis
df_hr['experience_level'] = pd.cut(df_hr['experience_years'], bins =[0, 2, 5, 10, 20], labels = ['Junior','Mid','Senior','Expert'],
                                   include_lowest = True) # [0,2], (2,5], (5,10], (10,20]

# Salary Bands
df_hr['salary_band'] = pd.qcut(df_hr['salary'], q = 4, labels = ['Low', 'Medium', 'High', 'Very High'])

# Employee Status Category
df_hr['attrition'] = df_hr['status'].apply(lambda x : 0 if x == 'Active' else 1) # Helps calculate attrition rate easily
```

### 8) Data Loading to SQL Server

```python
# Load the data into sql server
from sqlalchemy import create_engine, text

server = 'Kiran\\SQLEXPRESS'
database = 'hr_analytics'

engine = create_engine(f"mssql+pyodbc://@{server}/{database}"
                      "?driver=ODBC+Driver+17+for+SQL+Server"
                      "&trusted_connection=yes"
                      )

with engine.begin() as conn:
    conn.execute(text('''IF OBJECT_ID('dbo.employee_analytics', 'U') IS NOT NULL
                            TRUNCATE TABLE dbo.employee_analytics;''')) # it keeps the table structure and deletes the data

df_hr.to_sql("employee_analytics", con = engine, index = False, chunksize = 50000, if_exists = 'append') # now we append the data to the empty table
```
